#### *Earlier Industrial Revolutions and the transition to Industry 4.0*

Four important moments in which industry changes revolutionized the economic sector, and overall the way of living, are studied in history. The content in this section is based on *Genesis of the revolutionary transition to industry 4.0 in the 21st century and overview of previous industrial revolutions* by U. A. Pozdnyakova [1]. 

The first Industrial Revolution took place from the late-18th century to the early 19th century. The production process changed with the introduction of steam as a source of energy, which led to the transition from manual labor to machine labor. The outcome was an industrialized society by the end of the 19th century. Following this era, in the Second Industrial Revolution (late-19th to early-20th century), the most notable event was the replacement of steam by electricity as a source of energy. Important inventions occured during this time (e.g., the incandescen-filament lamp was created by Thomas Edison in 1879), but the one that achieved notorious importance was the three-phase power station invented by Mikhail Dolivo-Dobrovolsky in 1891, as it allowed a wide incorporation of electricity in the industrial processes. These changes caused the need for the development of new information networks, that later gave rise to relevant communication technologies such as the telephone (1876), and the radio (1896). Mass production, the expansion of advertisement, and the rising communication and information system, resulted in the emerging of brands as market leaders just as Coca-cola and Levi's. 

The Third Industrial Revolution began in the late-20th century. With the introduction of digital devices in the mid-20th century, automation started to scale rapidly, producing a shift in the focus of profitability from the production process to the design and development stages, causing a decrease in the number of blue-collar workers. Alvin Toffler, a contemporary philosopher, described the rise of this new economy as *a transition to society that is based on knowledge*. A significant change in the business models resulted in the replacement of traditional centralized economies by distributed structures, and renewable sources of energy appeared. 

The ongoing industrial transformation was introduced with a new term coined in Germany in 2012: 'Industry 4.0'. The fundamental idea of this new era is the integration of more computarized systems into the industry. Internet of Things emerges to improve interconnection and facilitate the production process by immediately adapting to the customers' needs, which allows mass production to be flexible and customizable. Similarly to the previous revolutions, Industry 4.0 will impact the current employment status by increasing the number of machine-driven operations, but now (and contrary to the Third Industrial Revolution) the automation stage might no longer need human interfaces at the operational stage.  

#### *The development of digital economy*

As described in the *Digital Economy Report* published by the United Nations Conference on Trade and Development (UNCTAD) in 2019 [2], the term *digital economy* was first used in the mid-1990's but its definition has evolved over the years. At the beginning, the internet was the center of the discussion, but in the past few years, digital economy has been associated more frequently with emerging technologies, being some of the most notorious AI, IoT, cloud computing, big data analytics and 3D printing. According to the same report, digital economy can be represented by breaking it down into three fundamental levels. The core corresponds to the digital (IT and ICT) sector, which holds hardware manufacture, software and IT consulting, information services, and telecommunications. The next layer (that is built on top of the previous one) is a narrow scope of digital economy where digital services and platform economy can be found; finally, the last level is a broader scope that also comprises e-Business, e-Commerce, Industry 4.0, precision agriculture, and algorithm economy. They also include sharing economy and gig economy in-between these two last levels. 

To emphasize the complexity of finding a consistent definition of digital economy, the report refers to various authors and their specific ways to conceptualize it. One of them describes digital economy as *that part of economic output derived solely or primarily from digital technologies with a business model based on digital goods or services*, and another one defines it as *encompassing all the ways in which digital technologies are diffusing into the economy*.

#### *Information technology in economic crises*

In *Information communication technology spending in (2008-) economic crisis*, Katarina Rojko identifies three ways in which the innovation adoption process can be affected by an economic crisis, which are: a slowed-down innovation diffusion process and a lower total revenue since buyers have fewer resources, postponed adoption by the majority, and faster adoption because in some cases the crisis can also have the opposite effect (e.g. innovations enabling decrease of costs) [3]. The same study shows how there has been a correlation between the gross domestic product (GDP) and the growth of IT spending during crisis periods in terms of individual regions and countries, depending on economy performance. 

Determining the impact of financial crises on matters related to information technology is a complex task that can be achieved only by looking at different indicators at once, which is exactly what the study *Information and Communication Technologies Developments in European Union during the Financial Crisis* explores. Their findings show that  during the crisis several indicators of the Information-Communication Technology (ICT) sector,  in  particular  those  related  to wide-areas services  and  networks,  presented  a significant increment, which led to the conclusion that the ICT  sector  in  the European Union  amortized  the  influences of the crisis. Nevertheless, there were still signs  like  the  negative  growth  of  ICT  revenues  and  ICT  net-work  investments,  which  indicate  that  the  developments in ICT will most likely be affected by the crisis in the subsequent years [4]. 


#### *The next Production Revolution*

The next Production Revolution is highly focused on three essential components: digital technologies (e.g. 3D printing, and IoT) new materials (e.g. bio- and nano-based), and new processes (e.g. data-driven processes, and AI) as statated in the report *The next production revolution: A report for the G20* prepared by the Organization for Economic Co-operation and Development
(OECD) in 2017 [5]. According to this report, some of the channels through which productivity can be affected, include the following: simulation, data-driven approaches, and the use of sensors can reduce significantly time and cost of production by predicting maintenance needs, reducing machine downtime and lowering the need of inventories as production can satisfy current rather than projected demand; similarly, 3D-printing can remove the assembling stage by producing already-assembled mechanisms.

Other important aspects presented by the OECD concern the time this progress in production will take, the impacts on the current employment status, how the economy will be benefited, what are some automation limits, and the need for new policies to regulate the transformed processes. The first point discusses how the availability of technology is not the only factor that determines the time of adoption of a new production model, as organizations opening channels for innovations is also crucial.  
About the emplyment status and the overall economy, the ways in which they will be benefited can be summarized with: lower prices of output, higher workers' wages, or higher profits. Regarding some of the limits of automation, the OECD considers the following traits to be still unlikely replicable by machines: creative intelligence, social intelligence, and common sense. Lastly, the OECD addresses the need of policies to monitor and prepare for a foreseeable situation in which income distribution will be highly unequal due to productivity-raising technologies.

1. **What were some of the most important events during the First Industrial Revolution?**

It is possible to distinguish some remarkable events which are closely related to each other: the use of steam and coal as sources of energy, the beginning of large-scale production, and the development of railroads. 

2. **What were some of the most important events during the Second Industrial Revolution?**

Some of the most relevant findings in the Second Industrial Revolution were the replacement of steam with electricity which led to the beginning of mass production, and the construction of new communication and information systems.

3. **What were some of the most important events during the Third Industrial Revolution?**

The most important aspects of this revolution are centered in automation and the production of computer systems and digital devices. 

4. **What characterizes the Industry 4.0?**

Industry 4.0 is characterized by an increment of processes automated with the use of emerging technologies and practices, such as IoT and Artificial Intelligence.

5. **How has the term digital economy changed over the years?**

In the 90's it was mostly related to the Internet but in recent years, with the introduction of new technologies and new economy models the term has become a center of discussion and debate as it is hard to give digital economy a consistent definition.

The infraestructure that can be considered the basis of the digital economy holds hardware manufacture, software and IT consulting, information services, and telecommunications.

7. **What is understood by digital transformation?**

Digital transformation can be understood in a general way as any economic output that is derived primarly from digital technologies.

8. **What are some technologies that could help the next Production Revolution occur?**

Technologies associated with the Internet of Things, Artificial Intelligence, as well as nano-materials and data-driven processes.

9. **What sectors might be affected by the next Production Revolution?**

The economic sector is the one that could be affected the most, considering aspects such as the current employment status. Although, this does not leave out any other sectors that are as well prone to be affected.

The cost and time of production can significantly decrease with the use of digital technologies and data-driven processes or the use of nano-materials.

There are many ways in which these processes can be affected, but the two plain possibilities are that either people will spend more on new technologies or that they will cease or lower their spendings on those things, depending on the context of the crisis.
